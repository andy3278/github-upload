```python
import numpy as np 
import matplotlib.pyplot as plt
import math
```

```python
x = np.linspace(0,10,10)
y = 2**x
```

```python
N = np.linspace(0,10,10)
N_square = N**2
two_power_N = y
N_factorial = []
```

```python
for w in N :
    w = math.factorial(int(w))
    N_factorial.append(w)
del N_factorial[10:200]
N_factorial
    
```

```python

plt.axis([0, 10, 0, 1000])
plt.plot(x, N, '-', label = "linear")
plt.plot(x, N_square,'-', label = "Square")
plt.plot(x, two_power_N, '-', label ="2^N")
plt.plot(x, N_factorial, '-', label = "N!")
```
